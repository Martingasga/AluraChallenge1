### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

## Análisis de datos

### 1.   Análisis de facturación





```python
tiendas = [tienda1, tienda2, tienda3, tienda4]
ingresos = {}

#Función para calcular la suma de cada columna de precios
def suma_precios(tienda):
    suma = float(tienda["Precio"].sum())
    #print(f"Suma total de precios: {suma}")
    return suma
```

```python
suma_precios1 = tienda1["Precio"].sum()
print(f"Ingreso total tienda 1: {suma_precios1}")
suma_precios2 = tienda2["Precio"].sum()
print(f"Ingreso total tienda 2: {suma_precios2}")
suma_precios3 = tienda3["Precio"].sum()
print(f"Ingreso total tienda 3: {suma_precios3}")
suma_precios4 = tienda4["Precio"].sum()
print(f"Ingreso total tienda 4: {suma_precios4}")
```

### 2. Ventas por categoría

```python
#Ventas por categoría - Cantidad de productos vendidos por tipo
ventas_categoria_tienda1 = tienda1.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)

#Mostrar resultados
print("\nVentas por categoría - Tienda 1:")
print(ventas_categoria_tienda1)
print(f"Total de productos vendidos en Tienda 1: {ventas_categoria_tienda1.sum()}")

print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_tienda2)
print(f"Total de productos vendidos en Tienda 2: {ventas_categoria_tienda2.sum()}")

print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_tienda3)
print(f"Total de productos vendidos en Tienda 3: {ventas_categoria_tienda3.sum()}")

print("\nVentas por categoría - Tienda 4:")
print(ventas_categoria_tienda4)
print(f"Total de productos vendidos en Tienda 4: {ventas_categoria_tienda4.sum()}")
```

### 3. Calificación promedio de la tienda


```python
#Calcular valoración promedio por tienda
valoracion_promedio_tienda1 = float(tienda1["Calificación"].mean())
valoracion_promedio_tienda2 = float(tienda2["Calificación"].mean())
valoracion_promedio_tienda3 = float(tienda3["Calificación"].mean())
valoracion_promedio_tienda4 = float(tienda4["Calificación"].mean())

print("\nValoración promedio por tienda:")
print(f"Tienda 1: {valoracion_promedio_tienda1:.2f} estrellas")
print(f"Tienda 2: {valoracion_promedio_tienda2:.2f} estrellas")
print(f"Tienda 3: {valoracion_promedio_tienda3:.2f} estrellas")
print(f"Tienda 4: {valoracion_promedio_tienda4:.2f} estrellas")
```

### 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
```

```python
#Función para mostrar productos más y menos vendidos
def mostrar_mas_menos_vendidos(df, nombre_tienda):
    ventas_productos = df.groupby("Producto")["Precio"].count().sort_values(ascending=False)

    mas_vendido = ventas_productos.head(1)
    menos_vendido = ventas_productos.tail(1)

    print(f"\n{nombre_tienda}")
    print("Producto más vendido:")
    print(mas_vendido)
    print("Producto menos vendido:")
    print(menos_vendido)

    #Gráfico de los 5 más vendidos
    plt.figure(figsize=(8, 4))
    ventas_productos.head(5).plot(kind="bar", color='red')
    plt.title(f"{nombre_tienda} - Top 5 productos más vendidos")
    plt.ylabel("Cantidad vendida")
    plt.xlabel("Producto")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
mostrar_mas_menos_vendidos(tienda1, "Tienda 1")
mostrar_mas_menos_vendidos(tienda2, "Tienda 2")
mostrar_mas_menos_vendidos(tienda3, "Tienda 3")
mostrar_mas_menos_vendidos(tienda4, "Tienda 4")
```

### 5. Envío promedio por tienda

```python
#Calcular el costo de envío promedio por tienda
envio_promedio_tienda1 = tienda1["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()

print("\nCosto de envío promedio por tienda:")
print(f"Tienda 1: ${envio_promedio_tienda1:.2f}")
print(f"Tienda 2: ${envio_promedio_tienda2:.2f}")
print(f"Tienda 3: ${envio_promedio_tienda3:.2f}")
print(f"Tienda 4: ${envio_promedio_tienda4:.2f}")
```

Con base en todos los factores analizados, **la recomendación es vender la Tienda 4**.

Justificación:

1. **Menores ingresos**: Tienda 4 tiene los ingresos más bajos (1038375700.0), lo que ya representa una desventaja directa frente a las demás.

2. **Calificación solo aceptable**: Aunque no es la peor, su calificación tampoco la destaca frente a otras, especialmente si la comparamos con Tienda 3 que tiene menor ingreso pero mejor reputación.

3. **Poca diversidad en categorías clave**: Tiene baja participación en Instrumentos Musicales, lo cual limita su mercado.

4. **Sin ventaja competitiva fuerte en ventas**: Aunque su producto más vendido tuvo 62 ventas (cerca de la Tienda 2), no destaca significativamente. Además, su producto menos vendido tiene tan solo 33 ventas.

5. **Única ventaja**: costo de envío más bajo, pero esto no compensa el bajo ingreso total, que es un factor clave si se busca rentabilidad.

¿Por qué NO vender las otras tiendas?
* **Tienda 1**: Aunque tiene la peor calificación y mayor costo de envío, genera los ingresos más altos, lo cual la hace financieramente atractiva.

* **Tienda 2**: Buen balance entre ingresos, productos vendidos y calificación. Su producto más vendido lidera en unidades.

* **Tienda 3**: Tiene la mejor calificación de clientes, lo que es clave para sostenibilidad a largo plazo. Aunque sus ingresos son medios, hay potencial de crecimiento con mejor stock o logística.

El Sr. Juan debería vender la Tienda 4. Las otras tres tiendas tienen ventajas claras y estratégicas que justifican su permanencia.
